# 4. Выбор модели
Чтобы выбрать между линейной или линейная в логарифмах, нужно провести **PE тест**:
1. Оцениваем линейную и полулогарифмическую (или логарифмическую) модель регрессии:
   - Линейная модель: $ Y = X\beta + \varepsilon $
   - Полулогарифмическая модель: $ \log Y = X\beta + \varepsilon $
2. Оценим вспомогательные регрессии
   - Для линейной регрессии: $ Y = X\beta + \delta_{LIN} (\hat{Y} - \log \hat{Y}) + \varepsilon $
   - Для полулогарифмической регрессии: $ \log Y = X\beta + \delta_{LOG} (\hat{Y} - \exp(\log \hat{Y})) + \varepsilon $
3. Проверим гипотезы $ H_0: \delta_{LIN} = 0$  и  $H_0': \delta_{LOG} = 0 $.

* $\delta_{LIN}$ pvalue 0.251
* $\delta_{LOG}$ pvalue 0.143

как мы видим pvalue lin и log достаточно большие, поэтому нулевые гипотезы не отвергается и мы не можем понять, какая модель лучше. грустно. Теперь попробуем **тест бокса-кокса**:

1. gY = $\sqrt[n]{Y_1 \cdot Y_2 \cdots Y_n}$
2. $Y^* = \frac{Y}{gY}$
3. Оцениваем регрессии:
   - $Y^* = \alpha + \beta_1 X + \varepsilon $
   - $\ln{(Y^*)} = \alpha' + \beta_1' X + \varepsilon' $
4. $\chi^2 = \frac{n}{2}|\ln{\frac{RSS_3}{RSS_4}}|$
5. 
    - $H_0$: качество подгонки моделей одинаково
    - $H_1$: лучше та модель, у которой RSS меньше
6. При $\chi^2 > \chi^2_{1,a}$ то нулевая гипотеза отвергается, где $\chi^2_{1,0.001} = 6.63$, $\chi^2_{1,0.05} = 3.84$, $\chi^2_{1,0.01} = 2.71$

В нашем случае $\chi^2$ = 1.79 < 2.71. То есть нулевая гипотеза не отвергается даже при уровне значимости 10%. Тоже не можем сказать какая модель лучше.

Линейная не лучше ни полулогарифмической, ни линейной в логарифмах, поэтому ее выбирать не будем.
Чтобы выбрать между линейной в логарифмах и полулогорифмической надо посмотреть на $R^2$. У полулогарифмической модели значение выше (0.247>0.241) => она лучше => выберем ее и дальше будем работать с ней.

# 5. Различия в зарплатах

## Тест Чоу
- $H_0$ (однородность выборки):
  $\text{все } \beta_i \text{ равны между выборками, то есть, } \beta_{1}' = \beta_{1}'', \ldots, \beta_{k}' = \beta_{k}'' $

- $H_1$ (неоднородность выборки):
  $\text{существует хотя бы один } i: \beta_{i}' \neq \beta_{i}'' $

Здесь $\beta$ обозначает коэффициенты модели, где ' и " соответствуют первой и второй выборке соответственно.

Тестовая статистика теста Чоу:

$ \frac{RSS_{\text{comb}} - (RSS_1 + RSS_2) / k}{(RSS_1 + RSS_2) / (n - 2k)} $

где $RSS_{\text{comb}}$ - сумма квадратов остатков объединенной модели, $RSS_1$ и $RSS_2$ - суммы квадратов остатков для каждой из подвыборок, $k$ - количество факторов в модели, $n$ - общее количество наблюдений.

Видим что pvalue 6.73e-06, то есть $H_0$ отвергается, то есть разница между зарплатами мужчин и женщин присутствует.
## Тест через dummy переменные
- $H_0$: $\beta_i = 0$ при дамми переменных и комбинациях с дамми
- $H_1$: существует дамми или комбинация с дамми, у которой $\beta_i\neq0$
Дальше рассчитывается F-статистика и проверяются гипотезы
В нашем случае pvalue вышел таким же как и в тесте Чоу => $H_0$ отвергается и выборка неоднородна.

# 10. Гетероскедастичность

## Тест Уайта

- $ H_0: \sigma^2_i = \sigma^2;  \forall i $ однородность дисперсии для всех наблюдений

- $ H_1: \exists \ \sigma^2_i \neq \sigma^2_j  $ наличие хотя бы двух наблюдений с различной дисперсией

Рассматриваем следующую регрессионную модель для остатков:

$ e^2 = \alpha_1 + \sum_{l=2}^{k} \alpha_l X_l + \sum_{l=2}^{k} \beta_{l2} X_l^2 + \sum_{l=2}^{k} \sum_{j=2}^{l-1} \gamma_{lj} X_l X_j + u $

При справедливости нулевой гипотезы статистика также имеет $\chi^2$ - распределение с (m - 1) степенями свободы, где (m) - число оцениваемых коэффициентов в модели.

pvalue получилось 0.98 => $H_0$ не отвергается
f_pvalue = 0.98
Из этого можно сделать однозначный вывод, что гетероскедастичность отсутствует или не значима

## Тест Бройша - Пагана
- $ H_0 $: $ \sigma_i^2 = \sigma_j^2 \forall i \neq j $
- $ H_1: \exists i, j $: $ \sigma_i^2 \neq \sigma_j^2 $


Тестовая статистика:

$ {\sigma}^2 = {RSS/2} $

Модель для теста Бройша-Пагана:

$ \frac{e^2}{\hat{\sigma}^2} = \gamma_0 + \gamma_1 Z_1 + \ldots + \gamma_p Z_p + u $

pvalue 0.99 => $H_0$ отвергается

Соответсвенно результат тот же, гетероскедастичности не выявлено

# 11. Тест рамсея
Тест Рамсея применяется для проверки адекватности спецификации регрессионной модели. Он позволяет выявить пропущенные переменные, неправильную функциональную форму модели или корреляцию между объясняющими переменными и ошибками.

- $H_0$:  $\alpha=0$ - коэффиценты добавленных переменных равны нулю (Модель правильно специфицирована)
- $H_1$: $\alpha\neq0$ - есть проблемы со спецификацией.

1. Оцениваем исходную регрессионную модель.
2. Расширяем модель, добавив в неё степени оцененных значений зависимой переменной.
3. F-тест для проверки статистической значимости добавленных переменных.

Если добавленные переменные статистически значимы, это указывает на неправильную спецификацию исходной модели.

F-статистика рассчитывается на основе сравнения исходной модели и расширенной модели:
$ F = \frac{(RSS_0 - RSS_1) / (m-1)}{RSS_1 / (n - (k + m))} $
где $ RSS_0 $ — сумма квадратов остатков исходной модели, $ RSS_1 $ — сумма квадратов остатков расширенной модели, $ n $ — количество наблюдений, $ k $ — количество переменных в исходной модели, $ m-1  $ — количество добавленных переменных.

pvalue 0.261 => $H_0$ не отвергается. Модель адекватно специфицировано все супер супер супер.!!21111!!11!
